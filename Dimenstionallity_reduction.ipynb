Импорт библиотек

```python
!pip install umap-learn
import pandas as pd
import numpy as np
from umap import UMAP
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

Инициализация алгоритмов

```python
umap = UMAP()
tsne = TSNE(n_jobs=-1)
pca = PCA(n_components=2)
algorithms = {"umap": umap, "tsne": tsne, "pca": pca}
```

Загрузка датасетов

```python
mnist_df = pd.read_csv("https://www.openml.org/data/get_csv/52667/mnist_784.csv")
```

```python
!pip install dataget
import dataget
_, __, f_mnist_test, f_mnist_y_test = dataget.image.fashion_mnist().get()
f_mnist = f_mnist_test.reshape(-1, 28*28)
```

```python
coil_20_df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/coil20.csv')
```

```python
!pip install gensim
from gensim.models import Word2Vec
from gensim.downloader import load
```

```python
google_news_dataset = load('word2vec-google-news-300').vectors
```

```python
datasets = [("MNIST", mnist_df[:20000].drop('class', axis=1), mnist_df[:20000]['class']), ("Fashion MNIST", f_mnist, f_mnist_y_test), 
            ("COIL20", coil_20_df.drop('1025', axis=1), coil_20_df['1025']), ("GoogleNews", google_news_dataset[:20000], None)]
```

Функция отрисовки

```python
def scatter(data, classes, alg):
  plt.figure(figsize=(10, 10))
  embedding = alg.fit_transform(data)
  plt.scatter(embedding[:,0], embedding[:,1], cmap = 'Spectral', 
                 s = 3, c = classes)
  plt.show()
```

Графики

UMAP

```python
for ds in datasets:
  print(ds[0])
  scatter(ds[1], ds[2], algorithms['umap'])
```

TSNE

```python
for ds in datasets:
  print(ds[0])
  scatter(ds[1], ds[2], algorithms['tsne'])
```

PCA

```python
for ds in datasets:
  print(ds[0])
  scatter(ds[1], ds[2], algorithms['pca'])
```
